### Find the most influencial characters

Which character has appeared in the maximum number of books?

```python
characterBookMap = edgesFiltered.map(lambda x:x.split())\
                                .map(lambda x : (x[0],x[1:]))
# (x[0] = first element is character id, x[1:] = second element is list of book ids in which that character appears)
```

```python
characterBookMap.take(10)
```

```python
# How to get character name for given character id? Create a dictionary of (character id, character name)
def charParse(row):
    row = row.split(":")
    return (row[0][7:],row[1].strip())

characterLookup = characters.map(charParse).collectAsMap()
```

```python
characterLookup
```

```python
# For example, what is character name for character id 10?
characterLookup['10']
```

```python
# Sort characters by strength (i.e. maximum books count in which each character has appeared)

characterStrength = characterBookMap.mapValues(lambda x:len(x))\
                                    .map(lambda x: (characterLookup[x[0]],x[1]))\
                                    .reduceByKey(lambda x,y:x+y)\
                                    .sortBy(lambda x:-x[1])

# mapValues - what's books count or length of books list
# map - prepare tuple(character name, books count)
# reduceByKey - what's sum of books count for each character name key
# sortBy - finally sort by books count or strength
```

```python
characterStrength.take(10)
```

__Answer__ = Spider-man/Peter Parker is most influencing character, as it appeared in maximum number of times in books among all other characters in Marvel Social Universe.

### Building a co-occurence Network

Characters connected to the books where they appeare in

```python
# Prepare map of book id and associated character ids

bookCharacterMap = characterBookMap.flatMapValues(lambda x: x)\
                                    .map(lambda x : (x[1],x[0]))\
                                    .reduceByKey(lambda x,y:x+","+y)\
                                    .mapValues(lambda x:x.split(","))\
# flatMapValues - prepare multiple records of <character id, book id> in case of iterable
# map - reverse to <book id, character id>
# reduceByKey - for each book id key, reduce to comma separated character ids
# mapValues - connect book id to list of character ids
```

```python
# Books connected to Characters
bookCharacterMap.take(10)
```

```python
# Prepare every possible pair of characters from the characters list of book using itertools module

import itertools
cooccurenceMap = bookCharacterMap.flatMap(lambda x: list(itertools.combinations(x[1], 2)))
```

```python
cooccurenceMap.take(10)
```

Each RDD represents pair of characters which appered together once in a book

```python
# What's count (i.e. cooccurence strength) for each pair of two connecting characters

cooccurenceStrength = cooccurenceMap.map(lambda x: (x,1))\
                                    .reduceByKey(lambda x,y : x+y)
```

```python
cooccurenceStrength.take(10)
```

```python
# How characters are connected through edges strength

cooccurenceEdges = cooccurenceStrength\
                   .map(lambda x : (x[0][0],x[0][1],x[1]))
cooccurenceEdges.take(10)
```

Each RDD represents pair of characters and its edges strength

### Understanding and modeling relationships 


__Requirement = Find top 10 relationships of characters by strength__

```python
sortedCooccurence = cooccurenceEdges\
                    .sortBy(lambda x:-x[2])\
                    .map(lambda x: (characterLookup[x[0]],characterLookup[x[1]],x[2]))
```

```python
sortedCooccurence.take(10)
```

__Requirement = Find strongest relationships for particular character__

```python
sortedCooccurence.filter(lambda x: 'SPIDER-MAN/PETER PARKER' in x).take(10)
```

```python
sortedCooccurence.filter(lambda x: 'CAPTAIN AMERICA' in x).take(10)
```

__Get quick sense of distribution of numeric edge weights__

```python
sortedCooccurence.map(lambda x:x[2]).stats()
```

```python
# edge weight with the interval of 10 with frequency distribution
sortedCooccurence.map(lambda x:int(x[2]/10)).countByValue()
```

__Visualization to find most important groups__

```python
import networkx as nx
G=nx.Graph()
edges = sortedCooccurence.map(lambda x:(x[0],x[1],{'weight':1000/x[2]})).take(50)

G.add_edges_from(edges)
```

```python
import matplotlib.pyplot as plt
nx.draw_networkx(G, pos=nx.circular_layout(G))
plt.show()
```
